```python
Splunk Dashboards / Reports / PostProcess Optimization — Jupyter Notes
1) Types of Dashboards
✅ Classic Dashboard

Format: JSON

Used in: Classic Splunk Dashboard framework

✅ Dashboard Studio

Format: XML

Used in: Modern Splunk dashboard UI

2) Dashboard vs Report
✅ Dashboard

We can create multiple panels

We can add charts, tables, filters, tokens, etc.

✅ Report

Report = Saved Search

We cannot create “panels” inside a report

But we can:

schedule the report

create alert from report

add report output to dashboards

3) PostProcess (Dashboard Optimization)
✅ What is PostProcess?

PostProcess means:

run one base search

reuse it in multiple panels by applying small filtering searches

✅ Why PostProcess?

reduces number of searches

improves dashboard speed

reduces CPU load on Search Head

✅ Example Structure

Base Search

index=notable earliest=-24h
| stats count by status owner


PostProcess panel search

| search status="new"
| stats sum(count) as open_notables

4) Dashboards, Reports, Alerts (Concept)
✅ Dashboard

for visualization

multiple panels

interactive view (tokens, drilldowns)

✅ Report

saved search output

can be reused in dashboards

can be scheduled

✅ Alert

scheduled / real-time triggering

can send email, create notable, webhook, etc.

5) Saved Searches Notes

Reports are Saved Searches

We can display events in visualizations

We can schedule reports and also add them to dashboards

6) Naming Convention (Best Practice)
✅ Format

Group: Department Name → IT, HR, SOC

Object: reports / dashboards

Description: schedule / purpose

✅ Example Use Case

“I want to send the weekly sales report to my manager”

✅ Name:

Sales_Report_Weekly_Sales_Revenue

7) Dashboard Permissions (Sharing)
✅ Permissions window includes:

Dashboard Name

Owner

App

Display For: Owner / App / All apps

Permissions:

Read

Write

✅ Typical sharing settings:

Everyone: Read ✅

Admin: Write ✅

Power: Write ✅ (if required)
```
