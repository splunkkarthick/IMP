```
Professional Summary & Architecture
•	Experienced Splunk Consultant managing data ingestion, indexing, and search head clusters.
•	Architecture handles single-site or multi-site clusters based on business needs.
•	Daily ingestion rate varies; typically ranges from GBs to TBs per day.
•	Retention depends on compliance, typically 90-365 days.
•	Multi-site clusters use configs like:
•	[clustering]
•	mode = master
•	multisite = true
•	site = site1
•	available_sites = site1, site2
•	replication_factor = 3
•	search_factor = 2

License Violation
•	License usage calculated daily (midnight to midnight UTC).
•	Up to 5 warnings in rolling 30-day window allowed.
•	Exceeding limit for 5+ days causes violation — disables all searches except internal indexes.

Log Sources Onboarded
•	Syslog (RFC 3164/5424) for network/unix devices
•	Windows Event Logs (EVTX)
•	CSV, JSON logs from applications/APIs
•	Custom multi-line text logs (stack traces)
•	XML logs for specialized devices

Windows & Linux Onboarding - Key Steps

1.	Prepare SOP, inputs.conf, outputs.conf, serverclass.conf.
2.	Ensure ports 8089 (management) and 9997 (indexing) are open.
3.	Set correct file permissions (Windows: check ACLs, Linux: ls -l, chmod).
4.	Monitor forwarder logs (splunkd.log) for errors.
5.	Validate forwarder process (ps -ef | grep splunkd or Windows Services).
6.	Use deployment server to push configurations.

Troubleshooting Forwarder Not Forwarding

1.	Check forwarder process running (ps -ef | grep splunkd or splunk status).
2.	Verify port 9997 listening on forwarder (netstat -an | grep 9997).
3.	Confirm indexer receiving enabled on port 9997.
4.	Test connectivity: ping, telnet <indexer_ip> 9997, or nc -zv <indexer_ip> 9997.
5.	Check file permissions (ls -l /path/to/log).
6.	Validate logs are updating and monitored paths are correct.
7.	Confirm inputs.conf and outputs.conf correct.
8.	Check indexer disk space (df -h).
9.	Review forwarder logs for connection errors.
10.	Verify Linux ulimit (ulimit -n), set recommended 65535.

Validate Indexer Receiving Port 9997
•	Run /opt/splunk/bin/splunk show listen to confirm port listening.
•	Check inputs.conf for:
•	[splunktcp://9997]
•	disabled = 0
•	Confirm port open (netstat -an | grep 9997 or ss -tln | grep 9997).
•	Verify firewall rules allow port 9997 (iptables or firewall-cmd).
•	Test connectivity from forwarder using telnet or nc.

```
