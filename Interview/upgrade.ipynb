```python
1. What is your backup strategy before upgrading Splunk?

Answer:
We take full backups of /opt/splunk/etc and KV store backups of /opt/splunk/var/lib/splunk/kvstore. Backup types include full, incremental, and differential backups. Daily, weekly, and monthly backups are automated via cron jobs and stored on a NetBackup solution.

2. Which directories and files are critical to backup in a Splunk environment?

Answer:

Configuration files: /opt/splunk/etc

KV store data: /opt/splunk/var/lib/splunk/kvstore

Indexers data: /opt/splunk/var/lib/splunk and /opt/splunk/etc

Any custom scripts in /opt/splunk/bin

App local directories: /opt/splunk/etc/apps/local

3. How do you backup the KV Store before an upgrade?

Answer:
Use the built-in Splunk CLI command:

/opt/splunk/bin/splunk backup kvstore


This ensures KV Store data integrity during upgrades.

4. Describe the upgrade path for a Splunk cluster.

Answer:
Upgrade order for independent servers:

Heavy Forwarders

Deployment Servers

Deployer

Cluster Master (enable maintenance mode before upgrade)

Indexers and Search Heads

5. How do you enable maintenance mode on the Cluster Master before upgrading?

Answer:
Run:

/opt/splunk/bin/splunk enable maintenance-mode


This pauses cluster activities to avoid data inconsistencies during the upgrade.

6. What should you do if the upgrade fails?

Answer:
Contact the backup team to restore the latest backup of /opt/splunk/etc and KV store from the backup solution to roll back to a stable state.

7. What are the different backup types mentioned?

Answer:

Full Backup: Complete snapshot of files.

Incremental Backup: Changes since the last backup.

Differential Backup: Changes since the last full backup.

8. What steps do you take to ensure add-ons compatibility during upgrade?

Answer:
Verify add-ons compatibility, especially for Heavy Forwarders and Search Heads, by confirming all add-ons support the target Splunk version.

9. What common issues may occur with DB Connect add-ons after upgrade?

Answer:
Inputs may get disabled, credentials might become corrupted, requiring reconfiguration of DB Connect inputs and credentials.

10. How do you backup Splunk etc directory using CLI?

Answer:
Use the tar command:

tar -czvf /tmp/backup/splunk_backup_$(date +%Y%m%d).tar.gz /opt/splunk/etc
```
