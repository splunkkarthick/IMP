```python
To determine the configuration file precedence, Splunk will consider the configuration in global context or App/user context.

Global: Activities like indexing take place in a global context.
App/user: Some activities, like searching, take place in an app or user context. 

Precedence within global context  

1. System local directory(etc/system/local) -- highest priority
2. App local directories(etc/apps/app_name/local)
3. App default directories(etc/apps/app_name/default) 
4. System default directory(etc/system/default) -- lowest priority

Precedence within global context, indexer cluster peers only

1. Peer-app local directories -- highest priority
2. System local directory
3. App local directories
4. Peer-app default directories
5. App default directories
6. System default directory -- lowest priority

Precedence within app or user context

1. User directories for current user (etc/user/user_name/App_currently_worling/local) -- highest priority
2. App directories for currently running app (etc/Apps/App_currently_worling/local) -- (local, followed by default)
3. App directories for all other apps (local, followed by default) -- for exported settings only
4. System directories (etc/system/local) finally (etc/system/default) (local, followed by default) -- lowest priority

Note : if two or more Apps have conflicting settings, App directory name with highest ASCII order wins
```
